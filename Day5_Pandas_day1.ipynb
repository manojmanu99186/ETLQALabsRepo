# Filter ( where clause in SQL)

```python
# get the data only for deptno=10
# SQL: select * from emp where deptno=10

df[df['deptno'] == 10]

# or 
deptno_10 = df['deptno'] == 10
df[deptno_10]
```

```python
# get the data only for deptno=10 and salary > 2K
# SQL: select * from emp where deptno=10

fil1 = df['deptno']==10
fil2 = df['salary']>2000
df_ans = df[fil1 & fil2]
df_ans
```

```python
# get the data only for deptno=10 or salary > 2K
# SQL: select * from emp where deptno=10

fil1 = df['deptno']==10
fil2 = df['salary']>2000
df_ans = df[fil1 | fil2]
df_ans
```

```python
# get the data only for (deptno=10 and salary > 2K)  or deptno=30
# SQL: select * from emp where (deptno=10 and salary > 2000) or  deptno=30

fil1 = df['deptno']==10
fil2 = df['salary']>2000
fil3 = df['deptno']==30
df_ans = df[(fil1 & fil2)|fil3]
df_ans
```

# Grouping data ( similar to sql group by)

```python
# get me emp count in every depratment
# select deptno,count(*) form emp group by deptno
var1 = df.groupby("deptno")["eno"].count()
var1
```

```python
type(var1)
```

```python
var2 = var1.reset_index()
```

```python
type(var2)
```

```python
var2
```

```python
var2.rename(columns = {'eno':"count_emp"})
```

```python
# get me emp count in every depratment
# select deptno,count(*) form emp group by deptno,salary
var1 = df.groupby(["deptno","salary"])["eno"].count().reset_index()
var1.head()
var1.rename(columns = {'eno':"count_emp"}).head()
```

```python
# get me emp count in every depratment
# select deptno,count(*) form emp group by deptno,salary
df.groupby('deptno').agg({"eno":"count","salary":"min"}).reset_index()
```
