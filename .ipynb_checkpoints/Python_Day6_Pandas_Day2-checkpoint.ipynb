# Joins in Pandas( Merge function)

```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/emp.csv")
```

```python
df2 = pd.read_csv("data/salaries.csv")
```

```python
df1
```

```python
df2
```

```python
# Inner join 
# select * from emp as e inner join salaries as s on e.eno = s.eno
ans_df = df1.merge(df2,on='eno')

# or

ans_df = df1.merge(df2,on='eno',how='inner')
```

```python
ans_df
```

```python
# Key ( joining column names are diffrent in both dataframe)
df1 = pd.read_csv("data/emp.csv")
df2 = pd.read_csv("data/salaries.csv")
```

```python
df1
```

```python
df2
```

```python
# Inner join 
# select * from emp as e inner join salaries as s on e.eno = s.eid
ans_df = df1.merge(df2,left_on='eno',right_on='eid',how='inner')
```

```python
ans_df
```

```python
# Left outer join 
# select * from emp as e inner join salaries as s on e.eno = s.eid
ans_df = df1.merge(df2,left_on='eno',right_on='eid',how='left')
```

```python
ans_df
```

```python
# Right outer join 
# select * from emp as e inner join salaries as s on e.eno = s.eid
ans_df = df1.merge(df2,left_on='eno',right_on='eid',how='right')
```

```python
ans_df
```

```python
# Full outer join 
# select * from emp as e inner join salaries as s on e.eno = s.eid
ans_df = df1.merge(df2,left_on='eno',right_on='eid',how='outer')
```

```python
ans_df
```

# Set Operators in Pandas

```python
# Union, Union All, Intersection, Minus
# we need to have same column name, same dataypes, no. of columns

df1 = pd.read_csv("data/emp1.csv")
df2 = pd.read_csv("data/emp.csv")
```

```python
df1
```

```python
df2
```

```python
# Intersection ( finding common rows/data)

#  way 1: 

df1.merge(df2,how='inner')

# or 

pd.merge(df1,df2,how='inner')
```

```python
# Intersection ( finding common rows/data)

#  way 2: 
df1.merge(df2,how='outer',indicator='source')
```

```python
df1.merge(df2,how='outer',indicator='source').query("source =='both'").drop("source",axis=1)
```

```python
# Minus query ( df1 - df2)
df1.merge(df2,how='outer',indicator='source').query("source=='left_only'").drop("source",axis=1)
```

```python
# Minus query ( df2 - df1)
df1.merge(df2,how='outer',indicator='source').query("source=='right_only'").drop("source",axis=1)

# or 

df2.merge(df1,how='outer',indicator='source').query("source=='left_only'").drop("source",axis=1)
```

```python
# Another way of Intersection
df = df1.merge(df2,how='outer',indicator='source')
```

```python
df[df['source'] == 'both'][['eno','ename']]
```

```python
# Having clause equivalent

# SQL: select deptno,sum(salary) from emp group by deptno having sum(salary)>30000
df =  pd.read_csv("emp.csv")
```

```python
df_grped = df.groupby("deptno")['salary'].sum()
```

```python
df_grped
```

```python
status_df = df_grped.apply(lambda x:x>30000)
```

```python
ans = df_grped[status_df].reset_index()
```

```python
ans
```

```python
# Scenario 1:  Reconcille data between Source ( oracle) and Target(mysql)

query_source = """select * from products"""
df_source = pd.read_sql(query_source,oracle_conn)

query_target = """select * from product_output"""
df_target = pd.read_sql(query_target,mysql_conn)

# check if all the product_ids are available in target

df_ans = df_source[~df_source['product_id'].isin(df_target['product_id'])]
extra_count = df_ans['product_id'].count()
if extra_count == 0:
    print("Test Passed")
else:
    print("Test Failed")
        

df_ans.to_csv("data/extra_in_source.csv",index=False)
```
